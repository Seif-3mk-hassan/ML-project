```python
from torch import nn
import torch

class simpleNN(nn.Module):
    def __init__(self, input_dim):
        super(simpleNN, self).__init__()
        self.fc1 = nn.Linear(input_dim, 512)
        self.relu = nn.ReLU()
        self.fc2 = nn.Linear(512, 256)
        self.relu2 = nn.ReLU()
        self.dropout1 = nn.Dropout(0.2)
        self.fc3 = nn.Linear(256, 128)
        self.relu3 = nn.ReLU()
        self.dropout2 = nn.Dropout(0.2)
        self.fc4 = nn.Linear(128, 64)
        self.relu4 = nn.ReLU()
        self.dropout3 = nn.Dropout(0.2)
        self.fc5 = nn.Linear(64, 1)

    def forward(self, x):
        out = self.fc1(x)
        out = self.relu(out)
        out = self.fc2(out)
        out = self.relu2(out)
        out = self.dropout1(out)
        out = self.fc3(out)
        out = self.relu3(out)
        out = self.dropout2(out)
        out = self.fc4(out)
        out = self.relu4(out)
        out = self.dropout3(out)
        out = self.fc5(out)
        return out
    
input_dim = X_train_scaled.shape[1]
model = simpleNN(input_dim)
criterion = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.005)
epochs = 2000
X_train_tensor = torch.tensor(X_train_scaled, dtype=torch.float32)
y_train_tensor = torch.tensor(y_train.values, dtype=torch.float32).view(-1, 1)
X_test_tensor = torch.tensor(X_test_scaled, dtype=torch.float32)
y_test_tensor = torch.tensor(y_test.values, dtype=torch.float32).view(-1, 1)
model.to('cuda')
X_train_tensor = X_train_tensor.to('cuda')
y_train_tensor = y_train_tensor.to('cuda')
X_test_tensor = X_test_tensor.to('cuda')
y_test_tensor = y_test_tensor.to('cuda')
for epoch in range(epochs):
    model.train()
    optimizer.zero_grad()
    outputs = model(X_train_tensor)
    loss = criterion(outputs, y_train_tensor)
    loss.backward()
    optimizer.step()
    print(f"Epoch [{epoch+1}/{epochs}], Loss: {loss.item():.4f}")
model.eval()
with torch.no_grad():
    y_nn_pred = model(X_test_tensor).cpu().numpy()
mse_nn = mean_squared_error(y_test, y_nn_pred)
r2_nn = r2_score(y_test, y_nn_pred)
print(f"Neural Network Mean Squared Error: {mse_nn:.2f}")
print(f"Neural Network R^2 Score: {r2_nn:.2f}")
```

```python
"""
Stacking Regressor — Best 3 Models
====================================
Base learners  : XGBoost, RandomForest, GradientBoosting
Meta-learner   : Ridge (better than plain LinearRegression for stacking,
                 as it regularises the blend coefficients)
CV strategy    : 5-fold cross-validation with out-of-fold predictions so
                 the meta-learner never sees training-fold data (prevents leakage)
"""

# ── Imports ──────────────────────────────────────────────────────────────────
import numpy as np
import xgboost as xgb
from sklearn.ensemble import (
    RandomForestRegressor,
    GradientBoostingRegressor,
    StackingRegressor,
)
from sklearn.linear_model import Ridge
from sklearn.model_selection import cross_val_score, KFold
from sklearn.metrics import mean_squared_error, r2_score

# ── Base estimators (same hyper-params as notebook) ──────────────────────────
xgb_model = xgb.XGBRegressor(
    n_estimators=700,
    max_depth=6,
    learning_rate=0.05,
    tree_method="hist",
    device="cuda",       # remove / change to 'cpu' if no GPU
    random_state=42,
    verbosity=0,
)

rf_model = RandomForestRegressor(
    n_estimators=700,
    max_depth=6,
    random_state=42,
    n_jobs=-1,
)

gb_model = GradientBoostingRegressor(
    n_estimators=700,
    max_depth=6,
    learning_rate=0.05,
    random_state=42,
)

# ── Meta-learner ──────────────────────────────────────────────────────────────
# Ridge with cross-validated alpha gives more stable blending weights
# than plain LinearRegression, especially when base models are correlated.
meta_learner = Ridge(alpha=1.0)

# ── Stacking Regressor ────────────────────────────────────────────────────────
# cv=5  → each base model is trained on 4 folds and predicts on the 5th,
#          producing out-of-fold (OOF) predictions for the meta-learner.
#          This is the standard Wolpert stacking protocol — no data leakage.
# passthrough=False  → meta-learner sees only the 3 base-model predictions,
#          keeping the second level lean and interpretable.
stacking_model = StackingRegressor(
    estimators=[
        ("xgb", xgb_model),
        ("rf",  rf_model),
        ("gb",  gb_model),
    ],
    final_estimator=meta_learner,
    cv=5,
    passthrough=False,   # set True to also feed raw features to meta-learner
    n_jobs=-1,
)

# ── Fit & Evaluate ────────────────────────────────────────────────────────────
print("Training stacking regressor …")
stacking_model.fit(X_train_scaled, y_train)

y_stack_pred = stacking_model.predict(X_test_scaled)

mse_stack  = mean_squared_error(y_test, y_stack_pred)
rmse_stack = np.sqrt(mse_stack)
r2_stack   = r2_score(y_test, y_stack_pred)

print("\n── Stacking Regressor (XGB + RF + GB → Ridge) ──")
print(f"  MSE  : {mse_stack:.4f}")
print(f"  RMSE : {rmse_stack:.4f}")
print(f"  R²   : {r2_stack:.4f}")

# ── Cross-validated R² on the full training set (optional sanity check) ──────
cv = KFold(n_splits=5, shuffle=True, random_state=42)
cv_r2 = cross_val_score(stacking_model, X_train_scaled, y_train,
                         scoring="r2", cv=cv, n_jobs=-1)
print(f"\n  5-fold CV R² on train : {cv_r2.mean():.4f} ± {cv_r2.std():.4f}")

# ── Individual base-model comparison ─────────────────────────────────────────
print("\n── Individual model R² (for reference) ──")
for name, preds in [("XGBoost",          y_xgb_pred),
                    ("RandomForest",      y_rf_pred),
                    ("GradientBoosting",  y_gb_pred)]:
    print(f"  {name:<20} R² = {r2_score(y_test, preds):.4f}")

# ── Meta-learner coefficients ─────────────────────────────────────────────────
print("\n── Ridge meta-learner blend weights ──")
meta = stacking_model.final_estimator_
for (nm, _), coef in zip(stacking_model.estimators, meta.coef_):
    print(f"  {nm:<5}  weight = {coef:.4f}")
print(f"  intercept = {meta.intercept_:.4f}")
```
